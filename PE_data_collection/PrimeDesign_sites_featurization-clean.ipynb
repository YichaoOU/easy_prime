```python
from Bio import SeqUtils
def is_PE3b(ngRNA_seq,PrimeDesign_input):
    ngRNA_seq = ngRNA_seq[1:] # remove first G
    seqrev = revcomp(ngRNA_seq)
    if ngRNA_seq in PrimeDesign_input:
        return False
    if seqrev in PrimeDesign_input:
        return False
    return True    
def is_dPAM(PAM_seq, RTT, cut_offset=-3):
    seq = revcomp(RTT)
    fwd_search = SeqUtils.nt_search(seq, PAM_seq)
    flag = 1
    if len(fwd_search) > 1:
        if abs(cut_offset) in fwd_search:
            flag = 0
    return bool(flag)    
def get_sgRNA(seq):
    if len(seq) == 21:
        return seq[1:]
    if len(seq) == 20:
        return seq
    print ("wrong")
    return ""
out = []
for s,d in df.groupby('site'):
    gene = s.split()[0]
    mutation = s.split()[1:]
    mutation.remove("-")
    mutation = "_".join(mutation)
    sgRNA = d.sgRNA.tolist()[0]
    PBS = d.PBS.tolist()[0]
    RTT1 = d.RTT1.tolist()[0]
    RTT2 = d.RTT2.tolist()[0]
    ngRNA1 = d.ngRNA1.tolist()[0]
    ngRNA2 = d.ngRNA2.tolist()[0]
    PAM = d.PAM.tolist()[0]
    PrimeDesign_input = d.PrimeDesign_input.tolist()[0]
    
    for rtt in [RTT1,RTT2]:
        if rtt == "":
            continue
        for ngRNA in [ngRNA1,ngRNA2]:
            if ngRNA == "":
                continue
            PE3b_flag = is_PE3b(ngRNA,PrimeDesign_input)
            dPAM_flag = is_dPAM(PAM, rtt)
            tmp = df2[df2.is_dPAM==dPAM_flag]
            if gene in ['ATP7B','SCN5A']: # these two ngRNAs are actually PE3 
                PE3b_flag = False
            tmp = tmp[tmp.is_PE3b==PE3b_flag]
            tmp = tmp[tmp.Site==gene]
            
            if tmp.shape[0]==0:
                print ("wrong")
                print (gene,dPAM_flag,PE3b_flag)
            print (gene,PE3b_flag,dPAM_flag,tmp.shape)
            eff = tmp.eff.mean()
            #Site, reported_mutation,dPAM, type ,sgRNA, PBS, RTT, ngRNA
            out.append([gene,mutation,dPAM_flag,PE3b_flag,get_sgRNA(sgRNA),PBS,rtt,get_sgRNA(ngRNA),eff,PrimeDesign_input])
            
df3 = pd.DataFrame(out)
df3.columns = ['gene','target_mutation','is_dPAM','is_PE3b','sgRNA','PBS','RTT','ngRNA','eff','PrimeDesign_input']
df3.head()
```

```python
def get_ref_alt_align(s):
    mutation = s[s.find("(")+1:s.find(")")]
    if "+" in mutation:
        return [0,0,len(mutation)-1]
    if "-" in mutation:
        return [0,len(mutation)-1,0]
    return [int((len(mutation)-1)/2),0,0]    
df3[['N_mutation','N_deletion','N_insertion']]=df3.PrimeDesign_input.apply(lambda x:pd.Series(get_ref_alt_align(x)))
df3.head()
```

```python
def get_refseq_altseq(s):
    before = s[:s.find("(")]
    after = s[s.find(")")+1:]
    mutation = s[s.find("(")+1:s.find(")")]
    if "+" in mutation:
        mutation = mutation.replace("+","")
        return [before+after,before+mutation+after]
    if "-" in mutation:
        mutation = mutation.replace("-","")
        return [before+mutation+after,before+after]
    mutation = mutation.split("/")
    return [before+mutation[0]+after,before+mutation[1]+after]
df3[['refseq','altseq']]=df3.PrimeDesign_input.apply(lambda x:pd.Series(get_refseq_altseq(x)))
def get_cut_pos(ampseq,sgRNA):
    pos1 = ampseq.find(sgRNA)
    pos2 = ampseq.find(revcomp(sgRNA))
    if pos1>0:
        return pos1+16
    if pos2>0:
        return pos2+3
    return -1
def get_nick_position(r):
    sgRNA = r.sgRNA.upper()
    ngRNA = r.ngRNA.upper()
    cut_sgRNA1 = get_cut_pos(r.refseq,sgRNA)
    cut_sgRNA2 = get_cut_pos(r.altseq,sgRNA)
    cut_ngRNA1 = get_cut_pos(r.refseq,ngRNA)
    cut_ngRNA2 = get_cut_pos(r.altseq,ngRNA)   
    if cut_ngRNA1 > 0:
        o =  cut_ngRNA1 - cut_sgRNA1
    if cut_sgRNA2 > 0:
        o= cut_ngRNA2 - cut_sgRNA2
    o= cut_ngRNA2 - cut_sgRNA1
    return [cut_sgRNA1,cut_sgRNA2,cut_ngRNA1,cut_ngRNA2,o]
df3[['cut_sgRNA1','cut_sgRNA2','cut_ngRNA1','cut_ngRNA2',"cut_position"]] = df3.apply(lambda x:pd.Series(get_nick_position(x)),axis=1)
```

```python
# target_to_pegRNA,target_to_RTT5
def get_cut_pos2(ampseq,sgRNA):
    pos1 = ampseq.find(sgRNA)
    pos2 = ampseq.find(revcomp(sgRNA))
    if pos1>0:
        return pos1+16,1
    if pos2>0:
        return pos2+3,-1
    return -1,0

def get_target_pos(r):
    # not generic code, specific to this dataset
    s = r.PrimeDesign_input
    sgRNA = r.sgRNA.upper()
    mutation = s[s.find("(")+1:s.find(")")]
    cut_sgRNA1,flag1 = get_cut_pos2(s,sgRNA)
    cut_sgRNA2,flag2 = get_cut_pos2(r.refseq,sgRNA)
    if flag1 == 1:
        pos = s.find("(")-1
        pos = pos - cut_sgRNA1
    elif flag1 == -1:
        pos = s.find(")")+1
        pos = cut_sgRNA1 - pos
    else:
        
        if flag2 == 1:
            pos = s.find("(")
            pos = pos - cut_sgRNA2
        elif flag2 == -1:
            pos = s.find(")")-3
#             print (r.gene,r.target_mutation,cut_sgRNA2,pos)
            pos = cut_sgRNA2 - pos
        else:
            pos = -9999

    return pos,len(r.RTT)-r.N_insertion-pos
df3[['target_pos',"RTT_flank"]]  = df3.apply(lambda x:pd.Series(get_target_pos(x)),axis=1)
```
